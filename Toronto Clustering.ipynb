```python
import numpy as np
import pandas as pd
import matplotlib as mpl
```

```python
path="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"   #storing the url in a variable
data=pd.read_html(path)                     # making use of the pandas read_html function to read values
label=['Postcode','Borough','Neighbourhood']          # defining the labels
df=pd.DataFrame(data[0])                    # taking the list elements into a dataframe
df.columns=df.iloc[0]                       # defining the proper column names
```

```python
df=df.drop(0)                       # cleaning the data 
df=df[df.Borough!="Not assigned"]   # removing "not assigned" Borough values
df.head()
```

```python
dff=df.groupby(['Postcode','Borough'],as_index=False).agg('sum')         # grouping the data
display(dff)                                                             # desired dataframe
```

```python
dff.shape    # shape of the final dataframe
```

```python
path1="https://cocl.us/Geospatial_data"  # storing the path of the csv file in a variable
ldata=pd.read_csv(path1)                 # making use of pandas read_csv function to read the data into a dataframe
ldata.head()                             # displaying the stored data
```

```python
ldata.rename({'Postal Code':'Postcode'},axis=1,inplace=True)   #renaming the column so as to merge the data
ldata.head()
```

```python
fdata=pd.merge(dff,ldata,how="left", on="Postcode")              # merging the data
fdata.rename({'Postcode': 'Postal Code'}, axis=1, inplace=True)  # since the actual name of the column is Postal Code
display(fdata)                                                   # displaying the final dataset
```

GENERATING THE LATTITUDE AND LONGITUDE OF THE MAP OF TORONTO

```python
add='Toronto, TN'
geol=Nominatim(user_agent="TN_explorer")
location=geol.geocode(add)
latitude=location.latitude
longitude=location.longitude
print("The geographic coordinates of Toronto are {},{}:".format(latitude,longitude))
```

FINALLY, THE MAP OF TORONTO WITH THE NEIGHBOURHOODS SUPERIMPOSED ON THE TOP IN GREEN
 (Please zoom out the map to see the results. Also, if the map is not visible, please check out the image "map_img.jpg" as it is the output screenshot of cell 17. Thankyou !)

```python
map_toronto=folium.Map(location=[40.4642335,-80.6009058],zoom_start=11)
for lat,lng,borough,neighborhood in zip(fdata['Latitude'],fdata['Longitude'],fdata['Borough'],fdata['Neighbourhood']):
    label='{}, {}'.format(neighborhood,borough)
    label=folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
map_toronto
```
